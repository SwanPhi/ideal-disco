```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv("countries_1.csv")
us = data[data.country == "United States"]
china = data[data.country == "China"]
```

```python
plt.plot(us.year, us.population / 10**6) #absolute growth of population
plt.plot(china.year, china.population /10**6)
plt.title("Absolute growth of population in US and China")
plt.legend(["US", "China"])
plt.xlabel("year")
plt.ylabel("Population")
plt.show()
```

```python
us_growth = us.population / us.population.iloc[0] *100  #relative growth of population
china_growth = china.population / china.population.iloc[0] * 100
plt.plot(us.year, us_growth)
plt.plot(china.year, china_growth)
plt.title("Relative growth of population in US and China")
plt.legend(["US", "China"])
plt.xlabel("year")
plt.ylabel("population growth %")
plt.show()
```
