```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('obama.csv', parse_dates=['year_month']) #interpret as datetime rather than str
data.head()
```

```python
plt.plot(data.year_month, data.approve_percent, 'o', markersize=1.5, alpha=0.5) #little o as plot with small dots without line
data_mean = data.groupby('year_month').mean() #summerise data in each month
plt.plot(data_mean.index, data_mean.approve_percent, 'orange')
data_median = data.groupby('year_month').median()
plt.plot(data_median.index, data_median.approve_percent, 'green')
plt.legend(['Approval Percent','Mean','Median'])
plt.show()
```

```python
data_mean = data.groupby('year_month').mean() #summerise data in each month
plt.plot(data_mean.index, data_mean.approve_percent, 'red')
```

```python
#Want to show percentile as well

data_25 = data.groupby('year_month').quantile(0.25)
data_75 = data.groupby('year_month').quantile(0.75)
plt.plot(data_25.index, data_25.approve_percent, 'green')
plt.plot(data_median.index, data_median.approve_percent, 'orange')
plt.plot(data_75.index, data_75.approve_percent, 'blue')
plt.legend(['25th percentile','Median','75th percentile'])
plt.plot(data.year_month, data.approve_percent, 'o', markersize=1.5, alpha=0.5) #little o as plot with small dots without line

plt.show()
```
