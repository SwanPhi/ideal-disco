```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
data = pd.read_csv("countries_1.csv") 
data_2007 = data[data.year == 2007]
#data_2007.head()
gdp_in_billions = (data_2007.gdpPerCapita * data_2007.population)/10**9#"gdp per capita is per person."
plt.scatter(gdp_in_billions, data_2007.lifeExpectancy, 5)
plt.title("Relationship between GDP and Life Expectancy in 2007")
plt.xlabel("GDP(in billions)")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
data_2007.lifeExpectancy.corr(gdp_in_billions)  #less correlated
```

```python
plt.scatter(np.log10(gdp_in_billions), data_2007.lifeExpectancy)
plt.show()
```

```python
data_2007.lifeExpectancy.corr(np.log10(gdp_in_billions)) #more highly correlated
```

```python
plt.scatter(np.log10(gdp_in_billions), data_2007.lifeExpectancy, 5)
plt.title("Relationship between GDP in log scale and Life Expectancy in 2007")
plt.xlabel("GDP(log scale)")
plt.ylabel("Life Expectancy")
plt.show()
```

```python
print("GDP per Capita correlated with life expectancy")
print(np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy))
print("GDP correlated with life expectancy")
print(np.log10(gdp_in_billions).corr(data_2007.lifeExpectancy))
```

```python
#using GDP per person shows more highly correlated to life expectancy. 
#It is more important for the person to have a high GDP himself/herself rather than the person have a high GDP of his/her
#country in total.
```
